```python
import mrcfile
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# Load the electron density map file
with mrcfile.open('emd_22183.map') as mrc:
    map_data = mrc.data
```

```python
print(map_data.shape)
```

```python
xy = np.mean(map_data, axis=2)
print(xy.shape)
```

```python
print(xy)
```

```python
plt.imshow(xy, cmap='hot')
plt.show()
```

```python
xz = np.mean(map_data, axis=1)
print(xz.shape)
```

```python
plt.imshow(xz, cmap='hot', interpolation='nearest')
plt.show()
```

```python
yz = np.mean(map_data, axis=0)
print(yz.shape)
```

```python
plt.imshow(yz, cmap='hot', interpolation='nearest')
plt.show()
```
